```python
import folium
import pandas as pd

def createLabel(dfRow):
    labelList = []

    if dfRow["Antiquario"] > 0:
        labelList.append("Antiquario")
    elif dfRow["Collezionista"] > 0:
        labelList.append("Collezionista")
    elif dfRow["Galleria"] > 0:
        labelList.append("Galleria")
    elif dfRow["Asta"] > 0:
        labelList.append("Casa d'aste")
    
    if len(labelList) == 1:
        label = str(labelList[0])
    else: 
        label = '<br>'.join(labelList)
    return label

#create a map
rawData = pd.read_csv("./antiquari.csv")
m = folium.Map(location=[43, 11.624819877497147], zoom_start=6)

for idx, row in rawData.iterrows():
    tooltip = row["Nome"]
    new_line = '<br>'
    bold_start = '<strong>'
    bold_end = '</strong>'
    if str(row["Schede"]).startswith("http"):
        link='<a href="{0}" target="blank_">Scopri le opere</a>'.format(row["Schede"])
    else:
        link=""
    text = f'{bold_start}{row["Nome"]}{bold_end}{new_line} \
    {createLabel(row)}{new_line}\{new_line}\{link}'
    marker = folium.Marker(
        location=[row["lat"], row["lon"]],
        popup=text,
        tooltip=tooltip)
    marker.add_to(m)

m.save("./antiquari.html")
m
```
